 # handling missing value

```python
df['last review'] = pd.to_datetime(df['last review'],errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'review per month':0,'last review' :df['last review'].min()},inplace = True)
```

```python
df
```

```python
#remove null values from column NAME & host name
df.dropna(subset = ['NAME','host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
#delete columns which is not needed - licence and house_rules
df = df.drop(columns=['license','house_rules'],errors='ignore')
```

```python
df.head()
```

```python
#check columns are removed or not
df.info()
```

```python
#remove $ signs from columns (price) & (service fee)
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
#check
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

# visualiztion

# What is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title('Distribution of listing price')
plt.xlabel('price $')
plt.ylabel('fequency')
plt.show()
```

```python
the histogram shows a fairly even distribution of listing prices across differnet price range,
indicating no concentration of in listng in anyspecific price range.
the kde line helps visualize this even spread more clearly,confirming that the dataset contains listing with a wide variety of prices.
```

# How are rooms type is distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x='room type',data = df, color = 'hotpink')
plt.title('room type distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

# How are listing distributed across different neighborhoods?

```python
plt.figure(figsize = (12,8))
sns.countplot(y='neighbourhood group',data=df,color='skyblue',order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

# What is the realationship between price and room type?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type', data=df, palette='Set2')
plt.title('price vs. room type')
plt.xlabel('Room type')
plt.ylabel('Price($)')
plt.legend(title='Room Type')
plt.show()
```

# How are the numbers of review change over with time?

```python
df['last review']=pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.ylabel('Number of reviews')
plt.show()
```
